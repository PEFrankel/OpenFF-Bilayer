```python
import matplotlib.pyplot as plt 
import numpy as np 
import math

plt.figure(1)
# OpenFF:
x,y = np.loadtxt('a71_a75_POPC.xvg',comments=["@","#"],unpack=True)
t,d = np.loadtxt('After_DB_62_POPE.xvg',comments=["@","#"],unpack=True)
plt.plot(x,y, color='red', label='POPC')
plt.plot(t,d, color='blue', label='POPE')

# Compared to:
w,z = np.loadtxt('a71_a75_POPC_true.xvg',comments=["@","#"],unpack=True)
j,k = np.loadtxt('After_DB_62_POPE_control.xvg',comments=["@","#"],unpack=True)
plt.plot(w,z, color='red', label='POPC', linestyle='dashed')
plt.plot(j,k, color='blue', label='POPE', linestyle='dashed')
Control = linestyle='dashed'

plt.xlabel("Degrees")
plt.ylabel("Probability Density")
plt.legend(["POPC", "POPE", "Control"])
plt.title("After Double Bond for all structures | HCCH")
plt.xlim(-180, 180)
plt.xticks([-180,-120,-60,0,60,120,180])

plt.figure(2)
# OpenFF:
l,m = np.loadtxt('a72_a75_POPC.xvg',comments=["@","#"],unpack=True)
q,s = np.loadtxt('After_DB_63_POPE.xvg',comments=["@","#"],unpack=True)
plt.plot(l,m, color='red', label='POPC')
plt.plot(q,s, color='blue', label='POPE')

# Compared to:
o,p = np.loadtxt('a72_a75_POPC_true.xvg',comments=["@","#"],unpack=True)
a,b = np.loadtxt('After_DB_63_POPE_control.xvg',comments=["@","#"],unpack=True)
plt.plot(o,p, color='red', label='POPC', linestyle='dashed')
plt.plot(a,b, color='blue', label='POPE', linestyle='dashed')

plt.xlabel("Degrees")
plt.ylabel("Probability Density")
plt.legend(["POPC", "POPE", "Control"])
plt.title("After DB | HCCC")
plt.xlim(-180, 180)
plt.xticks([-180,-120,-60,0,60,120,180]);

# plt.figure(3)
# t,d = np.loadtxt('11-14.xvg',comments=["@","#"],unpack=True)
# j,k = np.loadtxt('11-14_true_control.xvg',comments=["@","#"],unpack=True)
# plt.plot(t,d, color='red', label='OpenFF')
# plt.plot(j,k, color='blue', label='MacRog')
# plt.xlabel("Degrees")
# plt.ylabel("Probability Density ($nm^{-1}$)")
# plt.legend()
# plt.title("C-C Dihedral: 11-14")
# plt.xlim(-180, 180)
# plt.xticks([-180,-120,-60,0,60,120,180]);
```
