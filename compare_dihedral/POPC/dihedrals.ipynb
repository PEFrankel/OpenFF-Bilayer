```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Double Bond
# OpenFF
phi = np.linspace(-180, 180, num=1000)
# U = 13.197694603020*(1+np.cos((2*180-phi)*(np.pi/180))) + -1.745204449790*(1+np.cos((1*180-phi)*(np.pi/180)))

U = (13.197694603020-1.745204449790)*(1+np.cos((3*360-phi)*(np.pi/180)))

# U = k*(1+np.cos(n*phi-phi0))

# peak at cos = 1

# MacRog
# phi = np.linspace(-np.pi, np.pi, num=1000)
# P = 2*(1+np.cos(1*0.23034322449481356-phi)) + 1*(1+np.cos(1*-0.030459563769216-phi))

plt.figure(1)
plt.plot(phi, U)
plt.xlabel("Degrees")
plt.ylabel("$V_d$ ($kJ mol^{-1}$)")
plt.title("Double Bond Dihedral Potential (POPC)")
# plt.plot(phi,U, color='red', label='OpenFF')
# plt.plot(phi,P, color='blue', label='MacRog')
plt.xlim(-180, 180)
plt.xticks([-180,-120,-60,0,60,120,180])
```
