```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Headgroup

# *******************************
# ***** OpenFF calculations *****
# *******************************
# (Type 2)

phi_rad = np.linspace(-np.pi, np.pi, num=1000)
U = -3.176451728097*(1+np.cos(3*phi_rad-0)) + -6.567761030559*(1+np.cos(2*phi_rad-0)) + 34.325785207797*(1+np.cos(1*phi_rad-0))

# U = k*(1+np.cos(n*phi-phi0))

# peak at cos = 1

# *******************************
# **** CHARMM36 calculations ****
# *******************************
# (Type 9)

V = 0.4184*(1+np.cos(2*phi_rad-np.pi)) + 0.4184*(3+np.cos(2*phi_rad-np.pi)) + 5.0208*(1+np.cos(1*phi_rad-np.pi))

# COPO

phi = np.linspace(-180, 180, num=1000)
plt.figure(1)
plt.plot(phi, U)
plt.plot(phi, V)
plt.xlabel("Degrees")
plt.ylabel("$V_d$ ($kJ mol^{-1}$)")
plt.title("Headgroup Dihedral Potential | COPO")
plt.plot(phi,U, color='red', label='OpenFF')
plt.plot(phi,V, color='blue', label='CHARMM36')
plt.legend()
plt.xlim(-180, 180)
plt.xticks([-180,-120,-60,0,60,120,180]);
```
